```python
import nflgame
import nflgame.player as playe
def week_number(game):
    return nflgame.sched.games[game.eid]['week']

def game_number(team, weekno):
    info = nflgame.sched.games[g.eid]
    number = 0
    for d in nflgame.sched.games.itervalues():
        y, t, w = d['year'], d['season_type'], d['week']
        h, a = d['home'], d['away']
        equal = [
            y == info['year'],
            t == info['season_type'],
            w <= weekno,
            team in (h, a),
        ]
        if not all(equal):
            continue
        number += 1
    return number
```
